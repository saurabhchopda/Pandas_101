# Import pandas library and input data from a csv file

```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

```python
df[['Name', 'HP', 'Attack']]
```

```python
df.iloc[0:3]
```

# Search for integer number [row,column] using iloc

```python
df.iloc[78,3]
```

# Print the [index,rows] of the "head" of the data

```python
df1 = df.head()
for index, rows in df1.iterrows():
    print(index, rows)
```

# Query the data to find information from specific column

```python
df.loc[df['Type 1'] == 'Water']
```

# Describe the statistical properties of the data

```python
df.describe()
```

# Sort the data of a specific property (column) in descending order

```python
df.sort_values('Speed', ascending = False)
```

# Add all the numerical properties of data to get a new feature

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

# Sort the newly added feature in descending order

```python
df.sort_values('Total', ascending = False).head()
```

```python
df.iloc[:, 4:9]
```

# Use the groupby function to perform analysis and query the data easily

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
```

```python
df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```

# Read data in chunks (Useful for very huge datasets)

```python
for dff in pd.read_csv('modified.csv', chunksize = 1):
    print(dff)
```
